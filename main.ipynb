### DATA PREPARING

```python
# Drop unnecessary columns (if needed)
columns_to_drop = ['Unnamed: 0']  # Add columns to be dropped
df = df.drop(columns=columns_to_drop, axis=1)
```

```python
# Convert 'Transaction_Date' and 'Date' columns to datetime format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Extract additional features from datetime columns
df['Transaction_Year'] = df['Transaction_Date'].dt.year
df['Transaction_Month'] = df['Transaction_Date'].dt.month
df['Transaction_Day'] = df['Transaction_Date'].dt.day
```

```python
# Convert categorical columns to numerical using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Gender', 'Location', 'Coupon_Status', 'Product_Category'], drop_first=True)
```

```python
# Scale numeric features if necessary
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_columns = df_encoded.select_dtypes(include=['float64']).columns
df_encoded[numeric_columns] = scaler.fit_transform(df_encoded[numeric_columns])
```

```python
from scipy.stats import mstats

df_encoded['Quantity'] = mstats.winsorize(df_encoded['Quantity'], limits=[0.05, 0.05])
```

### CLASSIFICATION MODELING

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Extract the target variable 'Location'
y = df['Location']

# Extract relevant features
features = df[['Gender', 'Tenure_Months', 'Product_Category', 'Quantity', 'Avg_Price', 
               'Delivery_Charges', 'Coupon_Status', 'GST', 'Offline_Spend', 'Online_Spend', 
               'Month', 'Discount_pct', 'Product_Description']]
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encode the 'Gender', 'Coupon_Status', and 'Product_Category' columns
label_encoder = LabelEncoder()
features['Gender'] = label_encoder.fit_transform(features['Gender'])
features['Coupon_Status'] = label_encoder.fit_transform(features['Coupon_Status'])
features['Product_Category'] = label_encoder.fit_transform(features['Product_Category'])
```

```python
# One-hot encode the 'Product_Description' column
features = pd.get_dummies(features, columns=['Product_Description'], drop_first=True)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_classifier.fit(X_train, y_train)
```

```python
rf_classifier.score(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Display classification report and confusion matrix
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

### CLUSTERING MODELING

```python
# Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Define features for clustering
features_for_clustering = df_encoded[['Quantity', 'Avg_Price', 'Offline_Spend', 'Online_Spend', 'Discount_pct']]
```

```python
# Standardize the features
scaler = StandardScaler()
features_for_clustering_scaled = scaler.fit_transform(features_for_clustering)
```

```python
# Determine the optimal number of clusters using the Elbow Method
wcss = []  # Within-Cluster Sum of Squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(features_for_clustering_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method graph
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

```python
# Based on the Elbow Method, choose the optimal number of clusters (K)
optimal_k = 3  
```

```python
# Apply K-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=42)
df_encoded['Cluster'] = kmeans.fit_predict(features_for_clustering_scaled)
```

```python
#Analyze the characteristics of each cluster
cluster_summary = df_encoded.groupby('Cluster').mean()
print(cluster_summary)
```

### ANALYSIS OF SALES AND REVUNUE

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year from the 'Date' column
df['YearMonth'] = df['Date'].dt.to_period('M')
```

```python
# Group by YearMonth and sum the 'Online_Spend' for total sales per month
monthly_sales = df.groupby('YearMonth')['Online_Spend'].sum().reset_index()
```

```python
# Plot the total sales per month
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'].dt.to_timestamp(), monthly_sales['Online_Spend'], marker='o')
plt.title('Total Sales per Month in the Past Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Identify the top products or product categories contributing the most to sales
top_categories = df.groupby('Product_Category')['Online_Spend'].sum().sort_values(ascending=False).head(5)
```

```python
# Plot the contribution of top categories to total sales
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue')
plt.title('Top 5 Product Categories Contributing to Sales')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```
